```python
def fix_year(x):
    if x.year > 1989:
        year = x.year - 100
    else:
        year = x.year
    return dt.date(year, x.month, x.day)
```

```python
data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_year)

data.head()
```
